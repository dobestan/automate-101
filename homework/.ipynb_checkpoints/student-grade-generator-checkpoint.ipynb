```python
import random

import pandas as pd
from faker import Faker
```

```python
faker = Faker('ko_KR')
```

```python
df = pd.DataFrame(columns=[
    "Name",
    "Address",
    "Phonenumber",
    "Math Score",
    "Science Score",
])
```

```python
for new_row_index in range(10000):
    phonenumber = "010" + "".join([str(random.randint(0, 9)) for x in range(8)])
    
    # 10% 확률에 대해서 대한민국 국가번호를 붙인 핸드폰 번호가 존재한다.
    if random.randint(0, 100) < 10:
        phonenumber = "8210" + phonenumber[3:]
    
    new_row = {
        'Name': faker.name(),
        'Address': faker.address(),
        'Math Score': random.randint(0, 100),
        'Science Score': random.randint(0, 100),
        'Phonenumber': phonenumber,
    }
    
    df.loc[new_row_index] = new_row
```

```python
# 과제 #


# 현재 10000명의 학생들에 대한 "이름", "주소", "휴대폰 번호", "수학 점수", "과학 점수" 데이터를 가지고 있습니다.
# 여러분은 이 데이터를 바탕으로 아래의 4개의 추가적인 열을 pandas를 이용하여 만드셔야 합니다. ( "전처리된 휴대폰 번호", "수학과목 학점", "과학과목 학점", "낙제 여부" )

# 다음과 같은 가정들이 있습니다.

# 1. 휴대폰 번호는 90% 의 학생들의 경우에는 "010XXXXXXXX" 포맷으로, 
# 10%의 학생들의 경우에는 국가번호가 포함된 "8210XXXXXXXX" 로 되어 있습니다.
# 최종적으로는 "010-XXXX-XXXX"로 변경되어야 합니다.

# 2. 점수에 따른 학점입니다.
#   2-1. ( 쉬운 버전; 절대평가 ): 80점 이상 A, 60점 이상 B, 그 외에는 C 입니다.
# 
#   2-2. ( 어려운 버전; 상대평가 ): 상위 20% A, 상위 20~40% B, 그 외에는 C 입니다.
#   * 이 내용에 대해서는 수업시간에 다루지 않았습니다. => 직접 검색하셔서 해결하셔야 합니다.

# 3. 낙제 여부는 수학 학점과 과학 학점이 모두 C인 경우에 낙제입니다.
# 낙제일 경우에는 True, 낙제가 아닐 경우에는 False 라고 해주시면 됩니다.


df["Preprocessed Phonenumber"] = "010-0000-0000"  # 전처리된 휴대폰 번호
df["Math Grade"] = "B"   # 수학과목 학점
df["Science Grade"] = "B"  # 과학과목 학점
df["Rejected"] = False  # 낙제 여부 
```

```python
import pandas as pd
```

```python
df = pd.read_excel("student.xlsx")
```

```python
def get_preprocessed_phonenumber(phonenumber):
    phonenumber = str(phonenumber)
    phonenumber = "0" + phonenumber[0:]
    if(phonenumber[1] == "8"):
        phonenumber=phonenumber[:1] + phonenumber[3:]
    phonenumber=phonenumber[:3] + '-' + phonenumber[3:]
    phonenumber=phonenumber[:8] + '-' + phonenumber[8:]
    phonenumber
    
    return phonenumber
```

```python
def get_preprocessed_math(math):
    strmath = str(math)
    intmath = int(strmath)
    if(intmath >= 80):
        math = "A"
    elif(intmath < 60):
        math = "C"
    else:
        math = "B"
    
    return math
```

```python
def get_preprocessed_science(science):
    strscience = str(science)
    intscience = int(strscience)
    if(intscience >= 80):
        science = "A"
    elif(intscience < 60):
        science = "C"
    else:
        science = "B"
    
    return science
```

```python
def my_test2(row):
    if (str(row['MathScore']) == "C") and (str(row['ScienceScore']) == "C"):
        return True
    else:
        return False
```

```python
df['Preprocessed Phonenumber'] = df.Phonenumber.apply(get_preprocessed_phonenumber)
```

```python
df['MathScore'] = df.MathScore.apply(get_preprocessed_math)
```

```python
df['ScienceScore'] = df.ScienceScore.apply(get_preprocessed_science)
```

```python
df['reject'] = True
```

```python
df['reject'] = df.apply(my_test2, axis=1)
```
